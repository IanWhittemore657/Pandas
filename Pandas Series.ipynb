# Series
 The differences from pandas and numPy array from a Series, is that a Series can have axis labels, meaning it can be indexed by a label, instead of just a number location. This means it does not need to hold numeric data.


```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
my_list = [10,20,30]
arr = np.array([10,20,30])
d = {'a':10,'b':20,'c':30}
```

```python
pd.Series(data=my_list)
```

```python
pd.Series(data=my_list,index=labels)
```

```python
pd.Series(my_list,labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(arr,labels)
```

```python
#using a dictionary
pd.Series(d)
```

```python
pd.Series(data=labels)
```

```python

pd.Series([sum,print,len])
```

```python
sales_Q1 = pd.Series(data=[600,450,200,150],index = ['South Africa', 'China','England', 'Brazil'])                                   
```

```python
sales_Q1
```

```python
sales_Q2 = pd.Series([610,500,210,100],index = ['South Africa', 'China','England', 'Japan'])                                   
```
